```python
import pandas as pd
import numpy as np
```

```python
path = "development.csv"
df = pd.read_csv(path)

df.head()
```

```python
df = df[["title", "article", "label"]]
df.head()
```

for now we are using only title, article and label. but first, we also have to handle the missing NaN texts because machine learning models cannot handle Nan texts

```python
df["title"] = df["title"].fillna("")
df["article"] = df["article"].fillna("")
```

with this method, if the title and article is missing, we replace it with empty string. this will help avoid crashes later

We will also merge the text fields for the models
we create a new column "text" = title and article

```python
df["text"] = df["title"] + " " + df["article"]
df[["text", "label"]].head()
```

we also have to define X and y. X is the input which the model leaves and y is the the output which the model has to predict. X is a text and y is its correct class. We should note that the number of rows sohuld match. (the shape)

```python
X = df["text"]      
y = df["label"]     

print(X.shape)
print(y.shape)
```

We should convert the texts into numbers. We will use TF-IDF for the conversion. We have to import it rom sklearn

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

Words should become numbers and teach article should become numeric vectors. 
fit_transform  learns te vocabulary and transforms text

```python
vectorizer = TfidfVectorizer(
    lowercase=True,
    stop_words="english",
    max_features=20000
)

X_tfidf = vectorizer.fit_transform(X)

X_tfidf
```

So, we must test the data before training. We will give %80 of the data for training and %20 for validation.
we will use strarify=y for to keep the class balance

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_val, y_train, y_val = train_test_split(
    X_tfidf,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
```

To sum up:
df: our raw data
text: title + article combined for later numericalization
X: our input text
y: labels
X_tfidf: numeric version of the text
X_train: our training data
X_val: validation data
